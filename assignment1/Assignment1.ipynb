### Train three_layer_neural_network on Make-Moons dataset ###

#### Import packages and visualize raw data

```python
import three_layer_neural_network as tlnn
import numpy as np
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt

X, y = tlnn.generate_data()
plt.scatter(X[:, 0], X[:, 1], s=40, c=y, cmap=plt.cm.Spectral)
plt.show()
```

#### Train with Tanh activation function

```python
model = tlnn.NeuralNetwork(nn_input_dim=2, nn_hidden_dim=3 , nn_output_dim=2, actFun_type='tanh')
model.fit_model(X,y)
model.visualize_decision_boundary(X,y)
```

#### Train with Sigmoid activation function

```python
model = tlnn.NeuralNetwork(nn_input_dim=2, nn_hidden_dim=3 , nn_output_dim=2, actFun_type='sigmoid')
model.fit_model(X,y)
model.visualize_decision_boundary(X,y)
```

#### Train with ReLu activation function

```python
model = tlnn.NeuralNetwork(nn_input_dim=2, nn_hidden_dim=3 , nn_output_dim=2, actFun_type='relu')
model.fit_model(X,y)
model.visualize_decision_boundary(X,y)
```
